```python
import json
import shutil
import os
import time
import re
import pandas as pd

combine = []
c = 1


timestamp = str(time.strftime("%Y%m%d_%H%M%S"))
samplefile = "clusters_202112.csv"
predictions_raw = "output_202112.tar.gz"
predictions = "predictions_" + timestamp + ".jsonl"
outputfile = "output_" + timestamp + ".csv"

shutil.unpack_archive(predictions_raw, "weekly_files/output/")
os.rename("weekly_files/output/output", predictions)          
```

```python
with open(predictions) as predFile:
    data = predFile.readlines()
    
```

```python
flattenedData = []
for i in data:
    e = json.loads(i)
#     print("jasgdjsdjca:::",e)
    for k,v in e.items():
#         Line=''
#         BeginOffset=""
#         EndOffset=""
#         Score=""
#         Text=""
#         Type=""
#         File=""
        if k == 'Entities': 
            if v ==[]:
                BeginOffset =""
                EndOffset=""
                Score=""
                Text=""
                Type=""
            elif v !=[]:
                for i in v:
                    for ek,ev in i.items():
                        if ek == "BeginOffset":
                            BeginOffset =ev
                        if ek == "EndOffset":
                            EndOffset =ev
                        if ek == "Score":
                            Score =ev
                        if ek == "Text":
                            Text =ev
                        if ek == "Type":
                            Type =ev
        elif k== "File":
            File = v
        elif k=="Line":
            Line = v
        
    
    flattenedData.append({"Line":Line,
                          "BeginOffset":BeginOffset,
                          "EndOffset":EndOffset,
                          "Score":Score,
                          "Text":Text,
                          "Type":Type,
                          "File":File})

                
        
        
        
```

```python
df = pd.DataFrame(flattenedData)
df = df.sort_values('Line')
df.head(5)
```

```python
# Start Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("S3_Analysis").getOrCreate()
```

```python
# Read in data from S3 Buckets
from pyspark import SparkFiles
url ="https://shivammittalbootcamp-bucket.s3.amazonaws.com/Other_STRBI_Table_No_02._Business_of_Scheduled_Banks_in_India.csv"
spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("Other_STRBI_Table_No_02._Business_of_Scheduled_Banks_in_India.csv"), sep=",", header=True)

# Show DataFrame
df.show()
```
